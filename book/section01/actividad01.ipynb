**El Despegue de "Cali Café"**
El lunes 26 de enero de 2026, Ana inaugura oficialmente su cafetería "Cali Café" en el sector de San Antonio, iniciando su sueño empresarial con un préstamo bancario de $2.500.000 que representa su capital semilla. Con la frescura de la mañana, cuando el termómetro marca $18^0C$, Ana invierte $850.000 en insumos de alta calidad, adquiriendo exactamente 45 bolsas de café del Valle, 12 cajas de leche y 60 panes artesanales. Para posicionar su marca, lanza una promoción del 10% de descuento (0,10) para los primeros 10 clientes del día. Al cierre de la jornada, Ana registra ventas brutas por $375.500,50 (incluyendo algunos centavos por ajustes de IVA), pero debe pagar de inmediato $120.000 por el canon de arrendamiento diario y registrar una pérdida de -$15.000 por una jarra de cristal que se quebró en la barra. Finalmente, se enfrenta a su primera cuenta por cobrar, ya que un cliente queda debiendo $43.750 tras prometer el pago para el día siguiente. Al apagar las luces, Ana observa sus recibos y se pregunta si los números reales de su operación reflejan el éxito que imagina.

```python

import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output

# Configuración de la historia y preguntas
preguntas = [
    {
        "texto": "1. Ana compró 45 bolsas de café y 12 cajas de leche. ¿A qué conjunto pertenecen estos datos de conteo?",
        "opciones": ["Irracionales (I)", "Naturales (N)", "Enteros Negativos (Z-)", "Fraccionarios"],
        "correcta": "Naturales (N)",
        "feedback": "¡Correcto! Usamos los Naturales para contar unidades completas."
    },
    {
        "texto": "2. La jarra que se quebró costó -$15.000. ¿Cómo clasificamos este valor en el balance?",
        "opciones": ["Natural (N)", "Irracional (I)", "Entero (Z)", "Dígito"],
        "correcta": "Entero (Z)",
        "feedback": "¡Muy bien! Los números negativos pertenecen a los Enteros y representan pérdidas o deudas."
    },
    {
        "texto": "3. El descuento del 10% equivale a 0,1. ¿En qué conjunto se ubica este decimal exacto?",
        "opciones": ["Naturales (N)", "Irracionales (I)", "Racionales (Q)", "Enteros (Z)"],
        "correcta": "Racionales (Q)",
        "feedback": "¡Exacto! Cualquier decimal que se puede escribir como fracción (1/10) es Racional."
    },
    {
        "texto": "4. Las ventas tuvieron centavos: $375.500,50. ¿Por qué este número NO es un Entero?",
        "opciones": ["Por ser ingreso", "Por tener parte decimal", "Por ser muy grande", "Por ser positivo"],
        "correcta": "Por tener parte decimal",
        "feedback": "¡Correcto! Los Enteros no admiten fracciones de unidad (centavos)."
    }
]

def renderizar_cuestionario(preguntas):
    for i, p in enumerate(preguntas):
        out = widgets.Output()
        radio = widgets.RadioButtons(
            options=p['opciones'],
            description='',
            disabled=False
        )
        boton = widgets.Button(description="Verificar Respuesta", button_style='info')
        
        display(Markdown(f"### {p['texto']}"))
        display(radio)
        display(boton)
        display(out)
        
        def on_click(b, r=radio, o=out, corr=p['correcta'], feed=p['feedback']):
            with o:
                clear_output()
                if r.value == corr:
                    print(f"✅ {feed}")
                else:
                    print(f"❌ Inténtalo de nuevo. Piensa en la definición del conjunto.")
        
        boton.on_click(on_click)
        display(Markdown("---"))

# Ejecutar el cuestionario
display(Markdown("# ☕ Cuestionario Interactivo: Cali Café"))
display(Markdown("Analiza los datos de la historia de Ana y clasifica los números reales según su uso en la administración."))
renderizar_cuestionario(preguntas)
```
