## 1. ¬øQu√© es un Porcentaje?
El porcentaje es una forma de expresar un n√∫mero como una **fracci√≥n de 100**. Se utiliza para comparar cantidades de forma estandarizada.

* **Etimolog√≠a:** Del lat√≠n *per centum* (por cada cien).
* **S√≠mbolo:** `%`
* **Definici√≥n:** El $n\%$ de una cantidad es la parte proporcional de esa cantidad si el total fuera 100.





## 2. La Triple Identidad
Un porcentaje se puede representar de tres maneras diferentes. Entender esto es la clave para calcularlos r√°pido:

| Porcentaje | Fracci√≥n | Decimal (Punto flotante) |
| :--- | :--- | :--- |
| **1%** | $1/100$ | $0.01$ |
| **10%** | $10/100 = 1/10$ | $0.1$ |
| **25%** | $25/100 = 1/4$ | $0.25$ |
| **50%** | $50/100 = 1/2$ | $0.5$ |
| **100%** | $100/100 = 1$ | $1.0$ |



## 3. C√≥mo calcular el Porcentaje de una Cantidad
Para hallar el porcentaje de un n√∫mero, la regla de oro es multiplicar la cantidad por el decimal correspondiente.

### F√≥rmula:
$$\text{Resultado} = \text{Cantidad} \times \left( \frac{\text{Porcentaje}}{100} \right)$$

### Ejemplo en "Cali Caf√©":
Si un cliente compra un caf√© de **$12.000** y hay un impuesto del **$19\%$** (IVA):
1. Convertimos el porcentaje a decimal: $19 / 100 = 0.19$
2. Multiplicamos: $12.000 \times 0.19 = 2.280$
3. **Total IVA:** $2.280.



## 4. Trucos de C√°lculo Mental
No siempre necesitas una calculadora. Usa estos atajos:

* **El 10% (La coma a la izquierda):** Para hallar el 10%, solo mueve la coma decimal un lugar a la izquierda.
  * *10% de 5.400 es 540.*
* **El 1% (Dos lugares):** Mueve la coma dos lugares a la izquierda.
  * *1% de 5.400 es 54.*
* **El 50% (La mitad):** Divide la cantidad por 2.
* **El 25% (La mitad de la mitad):** Divide por 4.



## 5. Aplicaciones en la Vida Real

1. **Finanzas:** Intereses bancarios y tasas de rentabilidad.
2. **Comercio:** Descuentos en temporadas de ofertas.
3. **Salud:** Concentraci√≥n de ingredientes en medicamentos o alimentos.
4. **Probabilidad:** "Hay un 80% de probabilidad de lluvia hoy".

Los porcentajes pertenecen al conjunto de los **N√∫meros Racionales ($\mathbb{Q}$)** porque siempre pueden expresarse como una fracci√≥n con denominador 100.
