# Exponentes en números reales

## Cálculo de los exponentes



```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from fractions import Fraction

# Forzamos el renderizado
%matplotlib inline

def actualizar_todo(porcentaje):
    # 1. Creamos la figura con un espacio extra abajo para el texto
    fig = plt.figure(figsize=(10, 5))
    ax1 = plt.subplot2grid((2, 2), (0, 0)) # Tarta
    ax2 = plt.subplot2grid((2, 2), (0, 1)) # Barra
    
    # --- Gráfico de Tarta ---
    ax1.pie([porcentaje, 100 - porcentaje], startangle=90, 
            colors=['#3498db', '#ecf0f1'],
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
    ax1.set_title(f"Representación: {porcentaje}%")

    # --- Gráfico de Barra ---
    ax2.barh([''], [100], color='#ecf0f1')
    ax2.barh([''], [porcentaje], color='#3498db')
    ax2.set_xlim(0, 100)
    ax2.set_title("Ubicación en la Recta")
    
    # --- Cálculos ---
    frac_simp = Fraction(porcentaje, 100)
    decimal = porcentaje / 100
    
    # --- Equivalencias Matemáticas (Dibujadas en la figura) ---
    # Creamos un tercer "eje" invisible para poner el texto
    ax_txt = plt.subplot2grid((2, 2), (1, 0), colspan=2)
    ax_txt.axis('off') # Escondemos los ejes
    
    # Construimos el texto con formato LaTeX para Matplotlib
    texto_lat = (
        f"Equivalencia: {porcentaje}%  $=$ "
        f"$\\frac{{{porcentaje}}}{{100}}$  $=$ "
        f"$\\mathbf{{\\frac{{{frac_simp.numerator}}}{{{frac_simp.denominator}}}}}$  $=$ "
        f"Decimal: $\\mathbf{{{decimal}}}$"
    )
    
    ax_txt.text(0.5, 0.5, texto_lat, size=18, ha='center', va='center', 
                bbox=dict(boxstyle="round", facecolor='#f8f9fa', edgecolor='#3498db'))

    plt.tight_layout()
    plt.show()

# Usamos interact para mostrar el slider y la figura
widgets.interact(actualizar_todo, porcentaje=widgets.IntSlider(
    value=50, min=0, max=100, step=1, 
    description='Ajustar %:', 
    layout={'width': '80%'}
));
```
