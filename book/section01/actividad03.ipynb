## C谩lculo de los exponentes



```python
# puedes aqu铆 operar


import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Markdown, clear_output
from fractions import Fraction

# Funci贸n para actualizar la visualizaci贸n
def actualizar_todo(porcentaje):
    # Crear la figura con 3 谩reas: Tarta, Barra y Texto/Fracci贸n
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # 1. Gr谩fico de Tarta
    ax1.pie([porcentaje, 100 - porcentaje], 
            startangle=90, 
            colors=['#3498db', '#ecf0f1'],
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
    ax1.set_title(f"Visualizaci贸n: {porcentaje}%")

    # 2. Gr谩fico de Barra (Fracci贸n Decimal)
    ax2.barh(['Total (100)'], [100], color='#ecf0f1')
    ax2.barh(['Total (100)'], [porcentaje], color='#3498db')
    ax2.set_xlim(0, 100)
    ax2.set_title("Ubicaci贸n en la Recta")

    plt.show()

    # 3. C谩lculo de Fracciones
    frac_original = f"\\frac{{{porcentaje}}}{{100}}"
    frac_simplificada = Fraction(porcentaje, 100)
    decimal = porcentaje / 100
    
    # Renderizado de f贸rmulas matem谩ticas
    display(Markdown(f"""
### Equivalencias Matem谩ticas:
| Porcentaje | Fracci贸n Base | Fracci贸n Simplificada | Decimal |
| :---: | :---: | :---: | :---: |
| **{porcentaje}%** | ${frac_original}$ | **$\\frac{{{frac_simplificada.numerator}}}{{{frac_simplificada.denominator}}}$** | **{decimal}** |
"""))

# Configuraci贸n del Slider
slider = widgets.IntSlider(
    value=50, min=0, max=100, step=1,
    description='Ajustar %:',
    style={'description_width': 'initial'},
    layout={'width': '80%'}
)

# Salida
out = widgets.interactive_output(actualizar_todo, {'porcentaje': slider})

display(Markdown("## М Simulador de Fracciones y Porcentajes"))
display(widgets.VBox([slider, out]))
```
