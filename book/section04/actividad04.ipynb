En Estados Unidos, los impuestos se aplican a nivel federal, estatal y local. Hay distintos impuestos aplicables segn el ingreso que generes o los bienes que poseas.

Un tipo de impuesto en Estados Unidos es:

Impuesto sobre la renta (Income tax)

El impuesto sobre la renta se aplica a individuos y empresas.

Existen tres niveles de imposicin:

Impuestos federales: Es el impuesto aplicado por el gobierno central.

Impuestos estatales: Los estados tienen su propio impuesto sobre la renta, y la tasa vara segn el estado. Algunos estados, como Texas o Florida, no cobran impuestos sobre la renta estatal.

Impuestos locales: En algunas ciudades, como Nueva York, tambin se aplican impuestos locales sobre la renta.

```python
def impuestos(income):
    # Resolucin analtica obtenida del sistema
    federal_rate = 0.3140
    state_rate = 0.1029

    federal = federal_rate * income
    state = state_rate * income
    return federal, state

# ------------------------------------
# INGRESO MANUAL POR EL USUARIO
# ------------------------------------
ingreso = float(input("Ingrese los ingresos gravables: "))
f, s = impuestos(ingreso)

print(f"\nPara ingresos de ${ingreso:,.0f}:")
print(f"Impuesto Federal: ${f:,.2f}")
print(f"Impuesto Estatal: ${s:,.2f}")
print(f"Total Impuestos : ${f+s:,.2f}")

# ------------------------------------
# TABLA DE 100 000 HASTA 1 000 000
# ------------------------------------
print("\nTabla de impuestos desde 100 000 hasta 1 000 000:\n")
print(f"{'Ingresos':>12}  {'Federal':>12}  {'Estatal':>12}  {'Total':>12}")
print("-"*55)

for I in range(100000, 1000001, 100000):
    f, s = impuestos(I)
    print(f"{I:12,.0f}  {f:12,.2f}  {s:12,.2f}  {f+s:12,.2f}")
```
