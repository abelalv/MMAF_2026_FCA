## 5. ¿Cuanto cuesta tener un hijo en Colombia?


# **¿Cuánto cuesta tener un hijo en Colombia?**
image.png
**Criar un hijo en Colombia puede costar entre 1.700 y 2.200 millones de pesos,** según un estudio de la Universidad EAN. No es una exageración: la cifra incluye alimentación, salud, educación, tecnología y demás necesidades básicas desde el nacimiento hasta los 18 años. El banco BBVA, por su parte, hizo una estimación más “conservadora”, que ronda los 1.349 millones de pesos.

Este panorama financiero podría explicar una de las principales tendencias demográficas del país: la drástica caída de la natalidad. **De acuerdo con el DANE, en 2024 se registraron solo 445.011 nacimientos, la cifra más baja desde que hay registros.** Esto representa una disminución del 33,4 % respecto a 2014 y del 13,7 % en comparación con 2023.

Entre las razones más comunes para no tener hijos se encuentran:

•	El alto costo de vida.

•	La dificultad para acceder a vivienda y servicios de salud.

•	La inestabilidad laboral.

•	La ausencia de redes de apoyo, especialmente para madres solteras.

Según el Estudio Sociodemográfico de Fecundidad (ESEF), solo el 16 % de las mujeres entre 15 y 49 años desea tener hijos en los próximos años.

Esto representa una disminución del 33,4 % respecto a 2014 y del 13,7 % en comparación con 2023.

**La educación: el gasto más alto**

Aunque la educación pública existe, **los gastos adicionales no son menores:** transporte, alimentación, útiles escolares y uniformes pueden sumar entre 250.000 y 400.000 pesos al mes. En colegios privados, las pensiones oscilan entre   500.000 y 5.000.000 mensuales, sin contar la educación superior, donde una carrera profesional en una universidad privada puede costar hasta $80 millones, más los gastos de sostenimiento.

**Otros costos claves en la crianza**

•	Primeros 3 años: pañales, vacunas, leche y guardería pueden representar 1.5 millones de pesos mensuales, es decir, 54 millones en total.

**•	Salud:** entre copagos, vacunas y atención privada, los gastos pueden llegar a 1 millón de pesos al mes.

**•	Tecnología y ocio:** entre tablets, celulares, internet y juguetes, se estiman entre 2 y 10 millones de pesos al año.

**•	Ropa y calzado:** los cambios de talla frecuentes pueden costar entre 800.000 y 1.5 millones de pesos cada seis meses.

Además, el tipo de familia también influye en los costos: una madre soltera puede gastar en promedio $398 millones, mientras que una familia extendida (con apoyo de abuelos, por ejemplo) puede llegar hasta los 1.121 millones de pesos.

**¿Y en otros países?**

Según la UNESCO, el costo promedio de crianza en América Latina también es elevado:

**•	México:** entre 200.000 y 250.000 dólares (equivalente a 800 – 1.000 millones de pesos)

**•	Argentina:** alrededor de 150.000 dólares.

Colombia se encuentra entre los países con mayor costo proporcional en relación al ingreso promedio. Y ese es precisamente el problema: la desigualdad entre ingresos y el gasto básico familiar. En 2024, **el ingreso mensual promedio de un hogar colombiano fue de 2.1 millones de pesos,** cifra que difícilmente permite cubrir los costos proyectados.

Tener un hijo es una decisión profundamente emocional, pero también debe ser una decisión consciente desde lo financiero. En el contexto económico actual de Colombia, prepararse es más importante que nunca.
