*Pronóstico de ventas*

## 6. Contexto

El **pronóstico de ventas** es el proceso de predecir lo que es probable que venda una empresa en un periodo futuro, normalmente medido en semanas, meses o trimestres. Estima los ingresos por ventas de acuerdos que ya están en curso o que se espera que entren en el *pipeline* de ventas. Un pronóstico de ventas confiable permite a las organizaciones operar con mayor claridad y control.

El objetivo principal del pronóstico de ventas es proporcionar una imagen clara de las ventas futuras y los ingresos futuros para que los líderes puedan tomar decisiones comerciales informadas. Los pronósticos ayudan a guiar áreas clave como cotizaciones, contratación, producción, gestión de inventario, planeación de ventas y estrategia.

La precisión de un pronóstico depende en gran medida de la calidad de los datos que lo respaldan. Las organizaciones de ventas con una sólida disciplina de datos —donde los registros de gestión de relaciones con los clientes (CRM) son actuales, completos y se actualizan constantemente— tienden a producir mejores pronósticos.

Crear un pronóstico requiere tanto estructura como criterio. El lado analítico se basa en estadísticas sobre ventas pasadas, velocidad de las transacciones, estacionalidad, análisis de tendencias y otras investigaciones de mercado. El lado intuitivo se basa en los *insights* de los representantes de ventas y gerentes que entienden el contexto específico de cada oportunidad.

A continuación encontrarás el código que genera ambas gráficas para la comparación:

```python
%pip install -q matplotlib numpy ipywidgets

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Markdown

# ─── Datos de ventas mensuales (60 meses / 5 años) ─────────────────────────
meses = np.arange(1, 61)
ventas = np.array([
    52, 48, 50, 58, 65, 72, 75, 73, 68, 62, 58, 54,  # Año 1
    53, 49, 51, 57, 64, 71, 74, 72, 69, 63, 59, 56,  # Año 2
    51, 47, 49, 59, 66, 73, 76, 74, 67, 61, 57, 53,  # Año 3
    54, 50, 52, 60, 67, 74, 77, 75, 66, 60, 56, 47,  # Año 4
    50, 46, 48, 56, 63, 70, 73, 71, 70, 65, 61, 65   # Año 5
])


def graficar_datos():
    """Genera la gráfica de dispersión de las ventas mensuales (puntos y unión)."""
    fig, ax = plt.subplots(figsize=(14, 5))
    ax.scatter(meses, ventas, color='steelblue', s=60, zorder=5, label='Ventas reales')
    ax.plot(meses, ventas, color='steelblue', alpha=0.5, linewidth=1.2)
    for corte in [12, 24, 36, 48]:
        ax.axvline(x=corte + 0.5, color='gray', linestyle='--', alpha=0.4)
    ax.set_xlabel('Mes (t)', fontsize=12)
    ax.set_ylabel('Ventas (millones COP)', fontsize=12)
    ax.set_title('Ventas mensuales – Empresa de electrodomésticos, Cali (60 meses)', fontsize=13)
    ax.set_xticks(np.arange(1, 61, 3))
    ax.set_xlim(0, 61)
    ax.set_ylim(30, 90)
    ax.grid(True, alpha=0.3)
    ax.legend()
    plt.tight_layout()
    plt.show()


def graficar_comparacion():
    """Compara la gráfica correcta con una versión que contiene errores en meses específicos."""
    ventas_error = ventas.copy()
    # Valores erróneos introducidos en los meses resaltados en rojo
    ventas_error[18] = 79  # Mes 19: correcto 74, error 79
    ventas_error[19] = 76  # Mes 20: correcto 72, error 76
    ventas_error[20] = 74  # Mes 21: correcto 69, error 74
    ventas_error[37] = 45  # Mes 38: correcto 50, error 45
    ventas_error[38] = 48  # Mes 39: correcto 52, error 48
    ventas_error[39] = 55  # Mes 40: correcto 60, error 55

    fig, axes = plt.subplots(2, 1, figsize=(14, 9), sharex=True)
    for ax, data, titulo in [
        (axes[0], ventas,       'Gráfica correcta (datos de la tabla)'),
        (axes[1], ventas_error, 'Gráfica de referencia (con errores)'),
    ]:
        ax.scatter(meses, data, color='steelblue', s=50, zorder=5)
        ax.plot(meses, data, color='steelblue', alpha=0.5, linewidth=1.2)
        ax.set_title(titulo, fontsize=12)
        ax.set_ylabel('Ventas (millones COP)')
        ax.grid(True, alpha=0.3)
        ax.set_ylim(30, 90)
        ax.set_xticks(np.arange(1, 61, 3))

    meses_error = [19, 20, 21, 38, 39, 40]
    for m in meses_error:
        axes[1].scatter(m, ventas_error[m - 1], color='red', s=100, zorder=6,
                        label='Punto con error' if m == 19 else '')
    axes[1].set_xlabel('Mes (t)', fontsize=12)
    axes[1].legend()
    plt.tight_layout()
    plt.show()


def ajustar_trig(funcion='sen', A=15.0, B=0.524, C=-2.09, D=62.0):
    """
    Grafica los datos reales y una función trigonométrica ajustada.

    Modelo:  V(t) = A · f(B·t + C) + D

    Parámetros
    ----------
    funcion : str   función a usar: 'sen', 'cos', 'tan', 'csc', 'sec', 'cot'
    A       : float amplitud (escala vertical de la oscilación)
    B       : float frecuencia angular  [B = 2π/T; período anual T=12 → B≈0.524]
    C       : float desplazamiento de fase (corrimiento horizontal)
    D       : float desplazamiento vertical (nivel medio de ventas)
    """
    trig_map = {
        'sen': np.sin,
        'cos': np.cos,
        'tan': np.tan,
        'csc': lambda x: 1.0 / np.sin(x),
        'sec': lambda x: 1.0 / np.cos(x),
        'cot': lambda x: np.cos(x) / np.sin(x),
    }
    f = trig_map.get(funcion)
    if f is None:
        print(f"Función '{funcion}' no reconocida. Opciones: sen, cos, tan, csc, sec, cot")
        return

    t_cont  = np.linspace(1, 60, 1000)
    V_cont  = A * f(B * t_cont + C) + D
    V_meses = A * f(B * meses  + C) + D
    V_cont  = np.where(np.abs(V_cont) > 200, np.nan, V_cont)
    error_promedio = float(np.mean(np.abs(ventas - V_meses)))

    fig, ax = plt.subplots(figsize=(14, 5))
    ax.scatter(meses, ventas, color='steelblue', s=60, zorder=5, label='Datos reales')
    ax.plot(meses, ventas, color='steelblue', alpha=0.3, linewidth=1)
    ax.plot(t_cont, V_cont, color='tomato', linewidth=2,
            label=f'V(t) = {A}·{funcion}({B:.3f}·t + {C:.3f}) + {D}')
    ax.set_xlabel('Mes (t)', fontsize=12)
    ax.set_ylabel('Ventas (millones COP)', fontsize=12)
    ax.set_title(f'Ajuste trigonométrico | Error promedio: {error_promedio:.2f} M COP', fontsize=12)
    ax.set_xlim(0, 61)
    ax.set_ylim(20, 100)
    ax.grid(True, alpha=0.3)
    ax.legend()
    plt.tight_layout()
    plt.show()
    display(Markdown(f'**Error promedio:** {error_promedio:.2f} millones COP'))


def analizar_errores(funcion='sen', A=15.0, B=0.524, C=-2.09, D=62.0):
    """Tabla y gráfica de error absoluto mes a mes entre datos reales y ajustados."""
    trig_map = {
        'sen': np.sin, 'cos': np.cos, 'tan': np.tan,
        'csc': lambda x: 1.0 / np.sin(x),
        'sec': lambda x: 1.0 / np.cos(x),
        'cot': lambda x: np.cos(x) / np.sin(x),
    }
    f = trig_map.get(funcion, np.sin)
    V_meses     = A * f(B * meses + C) + D
    errores     = ventas - V_meses
    errores_abs = np.abs(errores)

    lineas = [
        '| Mes | Ventas reales | V(t) ajustada | Diferencia | |Diferencia| |',
        '|:---:|:-------------:|:-------------:|:----------:|:-----------:|',
    ]
    for m, v, vt, e, ea in zip(meses, ventas, V_meses, errores, errores_abs):
        lineas.append(f'| {int(m)} | {v} | {vt:.2f} | {e:.2f} | {ea:.2f} |')
    display(Markdown('\n'.join(lineas)))

    idx_max = int(np.argmax(errores_abs))
    display(Markdown(
        f'\n**Error promedio:** {np.mean(errores_abs):.2f} M COP  \n'
        f'**Mes con mayor diferencia:** Mes {int(meses[idx_max])} '
        f'(|diferencia| = {errores_abs[idx_max]:.2f} M COP)'
    ))

    colores = ['tomato' if ea == np.max(errores_abs) else 'steelblue' for ea in errores_abs]
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.bar(meses, errores_abs, color=colores, alpha=0.8)
    ax.axhline(np.mean(errores_abs), color='orange', linestyle='--',
               label=f'Promedio: {np.mean(errores_abs):.2f}')
    ax.set_xlabel('Mes (t)')
    ax.set_ylabel('|Diferencia| (M COP)')
    ax.set_title('Error absoluto mes a mes')
    ax.legend()
    ax.grid(True, alpha=0.3, axis='y')
    plt.tight_layout()
    plt.show()
```
