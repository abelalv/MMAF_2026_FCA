## Notación matemática (leer antes de comenzar)
Para evitar errores de interpretación, usa estas convenciones durante toda la actividad:

1. $a_n$: término general de la sucesión en el período $n$.
2. $a_1$: primer término de la sucesión.
3. $d$: diferencia común (si la sucesión es aritmética).
4. $L$: límite propuesto de la sucesión (si se estudia convergencia).
5. $|a_n - L|$: distancia entre el término y el límite.
6. $n \geq N$: significa "para todo término desde $N$ en adelante".
7. $\lim_{n \to \infty} a_n = L$ significa que, cuando $n$ crece sin límite, los términos $a_n$ se acercan cada vez más a $L$.

### 8.4.2 Jugando con el computador
Una empresa monitorea el costo promedio por unidad de un proceso. El modelo del costo por período se define por recurrencia:

$$a_{n+1} = L + r(a_n - L), \quad a_1>0, \quad 0 < r $$

donde $L$ es el costo de equilibrio (en miles de COP) y $r$ controla la velocidad de convergencia.

Observa la siguiente tabla de valores para los primeros períodos:

```python
import ipywidgets as widgets
from IPython.display import clear_output, display


# Cierra widgets de ejecuciones anteriores sin tocar los nuevos.
_prev_ui = globals().get("_sim_ui")
_prev_out = globals().get("_sim_out")
if _prev_ui is not None:
    _prev_ui.close()
if _prev_out is not None:
    _prev_out.close()


def generar_sucesion(n_terminos, r, L, a1=400.0):
    terminos = [a1]
    for _ in range(n_terminos - 1):
        terminos.append(L + r * (terminos[-1] - L))
    return terminos


def imprimir_tabla(n_terminos, r, L, a1):
    terminos = generar_sucesion(n_terminos, r, L, a1)
    print(f"Sucesion: a_{{n+1}} = {L:.0f} + {r:.1f}*(a_n - {L:.0f})")
    print(f"a_1 = {a1:.1f}, L = {L:.0f}\n")
    print(f"{'n':>4} | {'a_n':>12} | {'|a_n - L|':>14}")
    print("-" * 37)
    for i, a in enumerate(terminos, start=1):
        print(f"{i:>4} | {a:>12.4f} | {abs(a - L):>14.4f}")


slider_terminos = widgets.IntSlider(
    value=20, min=5, max=50, step=1,
    description="Terminos (n):", continuous_update=False
)
slider_r = widgets.FloatSlider(
    value=0.5, min=0.1, max=0.9, step=0.1,
    description="Parametro r:", readout_format=".1f", continuous_update=False
)
slider_L = widgets.FloatSlider(
    value=250.0, min=50.0, max=500.0, step=10.0,
    description="Limite L:", readout_format=".0f", continuous_update=False
)
slider_a1 = widgets.FloatSlider(
    value=400.0, min=100.0, max=600.0, step=10.0,
    description="a_1 inicial:", readout_format=".1f", continuous_update=False
)

estado = widgets.HTML()
out = widgets.Output()


def actualizar(_=None):
    estado.value = (
        f"<b>r = {slider_r.value:.1f}</b> &nbsp;&nbsp; "
        f"<b>a_1 = {slider_a1.value:.1f}</b>"
    )
    with out:
        clear_output(wait=True)
        imprimir_tabla(
            slider_terminos.value,
            slider_r.value,
            slider_L.value,
            slider_a1.value,
        )


for control in (slider_terminos, slider_r, slider_L, slider_a1):
    control.observe(actualizar, names="value")

ui = widgets.VBox([slider_terminos, slider_r, slider_L, slider_a1, estado, out])
display(ui)
actualizar()

# Guarda referencias para poder cerrar esta interfaz en la siguiente ejecucion.
globals()["_sim_ui"] = ui
globals()["_sim_out"] = out
```
